```python
import numpy as np
import pandas as pd
from sklearn import tree
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import xgboost
```

```python
data = pd.read_csv("E:/database.csv")
```

```python
mag_type = data['Magnitude Type']
data = data.dropna(axis = 1,how='all')
data
```

```python
to_drop = ['Date',
           'Time',
         'ID',
         'Status',
         'Source',
           'Type',
         'Location Source',
         'Magnitude Source', 
          'Magnitude Error',
          'Magnitude Seismic Stations',
          'Azimuthal Gap',
          'Depth Seismic Stations',
          'Horizontal Distance',
          'Horizontal Error',
          'Root Mean Square',
          'Depth Error']
data.drop(to_drop, inplace=True, axis=1)
```

```python

data.dropna()
```

```python

data['Magnitude Type'] = data['Magnitude Type'].astype('category')
data['Magnitude Type'] = data['Magnitude Type'].cat.codes
data
```

```python
data
```

```python
Y =np.vstack(data["Magnitude"])

data.drop('Magnitude', inplace=True, axis=1)

# the loss is bigger with Magnitude Type
data.drop('Magnitude Type', inplace=True, axis=1)
```

```python
xgb = xgboost.XGBRegressor(n_estimators=100, learning_rate=0.08, gamma=0, subsample=0.9,
                           colsample_bytree=1, max_depth=7)
average = 0

for _ in range(100):
    X_train, X_test, y_train, y_test = train_test_split(data, Y, test_size=0.33, shuffle = True)
    
    xgb = xgb.fit(X_train, y_train)
    y_pred = xgb.predict(X_test)
    average += np.average((y_test - y_pred)**2)
average /= 100
average
```
